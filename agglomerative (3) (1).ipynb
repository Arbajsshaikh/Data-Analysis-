```python
import pandas as pd
```

```python
data = pd.read_csv('zoo.csv')
data.head()
```

```python
data.info()
```

```python
features = data.drop(['animal_name','class_type'], axis = 1)
```

```python
features.shape
```

```python
actual_labels = data['class_type']
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
my_cluster_model = AgglomerativeClustering(n_clusters=7)
```

```python
my_cluster_model.fit(features)
```

```python
my_cluster_model.labels_#(0,6)
```

```python
data['predicted_labels'] = my_cluster_model.labels_
data
```

```python
data[data['predicted_labels'] == 1]
```
